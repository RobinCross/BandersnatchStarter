```python

import altair as alt
from app.data import Database

# Create an instance of the Database class
db = Database()

# Get the DataFrame using the dataframe() method from data.py
df = db.dataframe()

# Create an interactive scatter plot of level vs. health
scatter_plot = alt.Chart(df).mark_circle(size=60).encode(
    x=alt.X('Level', title='Level'),
    y=alt.Y('Health', title='Health'),
    color=alt.Color('Rarity', title='Rarity'),
    tooltip=['Name', 'Level', 'Health', 'Rarity']
).properties(
    width=600,
    height=400,
    title='Level vs. Health by Rarity'
).interactive()

# Display the scatter plot
scatter_plot

# # Create an interactive histogram of monster levels
# histogram = alt.Chart(df).mark_bar().encode(
#     x=alt.X('Level', bin=True, title='Level'),
#     y=alt.Y('count()', title='Count'),
#     tooltip=['Level', 'count()']
# ).properties(
#     width=600,
#     height=400,
#     title='Distribution of Monster Levels'
# ).interactive()
#
# # Display the histogram
# #histogram
```
